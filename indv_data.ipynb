```python
import pandas as pd
import numpy as np
```

```python
datadir="D:/Machine Learning/Poverty_comp/data/"
a_train=pd.read_csv(datadir+'A_indiv_train.csv', index_col=['id','iid'])
b_train=pd.read_csv(datadir+'B_indiv_train.csv', index_col=['id','iid'])
c_train=pd.read_csv(datadir+'C_indiv_train.csv', index_col=['id','iid'])
```

```python
a_train.head()
```

```python
# Standardize features
def standardize(df, numeric_only=True):
    numeric = df.select_dtypes(include=['int64', 'float64'])
    
    # subtracy mean and divide by std
    #df[numeric.columns] = (numeric - numeric.mean()) / numeric.std()
    
    return df
    

def pre_process_data(df, enforce_cols=None):
    print("Input shape:\t{}".format(df.shape))
        

    df = standardize(df)
    print("After standardization {}".format(df.shape))
        
    # create dummy variables for categoricals
    df = pd.get_dummies(df)
    print("After converting categoricals:\t{}".format(df.shape))
    

    # match test set and training set columns
    if enforce_cols is not None:
        to_drop = np.setdiff1d(df.columns, enforce_cols)
        to_add = np.setdiff1d(enforce_cols, df.columns)

        df.drop(to_drop, axis=1, inplace=True)
        df = df.assign(**{c: 0 for c in to_add})
    
    df.fillna(0, inplace=True)
    
    return df
```

```python
print("Country A")
aX_train = pre_process_data(a_train)
#ay_train = np.ravel(a_train.poor)

print("\nCountry B")
bX_train = pre_process_data(b_train)
#by_train = np.ravel(b_train.poor)

print("\nCountry C")
cX_train = pre_process_data(c_train)
#cy_train = np.ravel(c_train.poor)
```

```python
# load test data
a_test = pd.read_csv(datadir+'A_indiv_test.csv', index_col=['id','iid'])
b_test = pd.read_csv(datadir+'B_indiv_test.csv', index_col=['id','iid'])
c_test = pd.read_csv(datadir+'C_indiv_test.csv', index_col=['id','iid'])
# process the test data
a_test = pre_process_data(a_test, enforce_cols=aX_train.columns)
b_test = pre_process_data(b_test, enforce_cols=bX_train.columns)
c_test = pre_process_data(c_test, enforce_cols=cX_train.columns)
```

```python
def aggregate(data, var):
    grouped=data.groupby(var).sum()
    return grouped
```

```python
datasets=[aX_train, bX_train, cX_train,a_test,b_test,c_test]

aX_train_i= aggregate(aX_train, 'id')
bX_train_i= aggregate(bX_train, 'id')
cX_train_i= aggregate(cX_train, 'id')
a_test_i= aggregate(a_test, 'id')
b_test_i= aggregate(b_test, 'id')
c_test_i= aggregate(c_test, 'id')
```
